## 2.1 Review the sample code provided below

What is this code doing at a high level?

Intent: Load a CSV file containing passenger counts for train services into a <span style="color:#B49AD2">pandas DataFrame</span> in a safe, loss‑free way, so that:

- No information is lost during import (even if the file is messy).
- Missing values are recognized consistently.
- Type decisions are deferred until you’ve inspected the data, preventing early mis-typing (like interpreting IDs as numbers or dates).

What object is created?

A <span style="color:#B49AD2">pandas DataFrame</span> (a 2‑dimensional table-like object with labeled rows and columns). Although Series is imported, <span style="color:#B49AD2">no Series is created</span> here.

What information about the data is being revealed?
At this stage: none of the contents are transformed; the DataFrame is read "as text", preserving the raw structure. However, the import options are chosen to reliably surface:

- Where data might be missing (NA, null, blanks, etc.),
- Whether columns need cleaning (since everything is read as “string” initially),
- Which columns are candidates for later conversion (e.g., dates, counts).

```python
import pandas as pd
from pandas import DataFrame as df, Series as se
# ^ Brings in the pandas library (the main tool for tabular data in Python).
#   Also imports DataFrame and Series with short aliases, but in this snippet
#   they aren't used directly. The main object we'll create is a DataFrame.

data_2023_2024 = pd.read_csv(
    "../DATA/train_service_passenger_counts.csv",  # <- path to the CSV file
    sep=",",                                    # <- fields are comma-separated
    dtype="string",                             # <- read EVERY column as string
                                                #    This avoids premature type guesses
                                                #    (e.g., IDs turned into numbers, 
                                                #    dates parsed incorrectly). It’s a
                                                #    “safe import,” especially helpful
                                                #    for messy or mixed-type columns.
    na_values=["", "NA", "N/A", "null", "NULL", "NaN"],  # <- treat these as missing
                                                         #    across all columns.
    keep_default_na=True,                        # <- ALSO use pandas' built-in list
                                                 #    of NaN representations
                                                 #    (e.g., 'None', '#N/A', etc.).
                                                 #    Combining both ensures robust
                                                 #    missing-data detection.
    low_memory=False                             # <- load in a single pass to get
                                                 #    more consistent dtype inference
                                                 #    metadata. This uses more memory
                                                 #    but is safer for large files
                                                 #    with mixed contents per column.
)
# RESULT: data_2023_2024 is a pandas DataFrame containing the CSV data, with
# all columns stored as string dtype and missing values unified as NaN.
```

Common pitfalls & how this code avoids them:

- ID columns losing leading zeros: Avoided by dtype="string".
- Mixed “null” representations breaking filters: Avoided by na_values + keep_default_na=True.
- Inconsistent dtype inference in big files: Mitigated by low_memory=False.
- Accidental date parsing: Avoided by not enabling automatic parsing at read time.

Summary:

- You are creating a DataFrame (not a Series) called data_2023_2024.
- The intent is to safely ingest a potentially messy CSV (preserving all raw text, unifying missing values, and deferring type conversions until you’ve explored the data).
- This approach helps you ask better questions of the data: e.g., “How do passenger counts vary by date/route/station?” and “Where are data quality issues?”

**High‑level intent**
The purpose of this command is to count how many missing values (NaN) occur in each column of your DataFrame.
It answers the question:

>“Where in the dataset do we have incomplete or missing information, and how much of it?”

This is one of the most important early steps in data cleaning, because missing data affects:

how you analyse the dataset
* which columns can safely be converted to numbers or dates
* whether rows can be grouped or aggregated
* whether some fields need imputation (filling)
* whether some columns are too incomplete to use


What object is being processed? 

* data_2023_2024 → a pandas DataFrame.
* isna() → creates a True/False mask showing whether each cell is missing.
* sum() → counts True values for each column.

So the output is a Series (a one‑dimensional labeled object) where:

* the index = column names
* the values = number of missing entries per column


**What the output means**

This tells you:

* ✔ Some columns have no missing values

* ⚠️ Group has some missing values

```python
data_2023_2024_copy_1 = data_2023_2024.copy()
```

**High‑level intent**

The purpose of this line is to **create a full, independent copy** of your original DataFrame.
This allows you to experiment, clean, transform, or restructure the data without risking changes to the original dataset.
Think of it as:

> “Make a safe working version of the data so I don’t accidentally damage the original.”


**What object is being created?**

A new **pandas DataFrame** named data_2023_2024_copy_1
It contains the **same data** and the **same structure** as data_2023_2024 at the moment of copying.

Why do we need .copy() instead of = ?

>✔ df_copy = df

* This does NOT create a real copy.
* It creates a reference (an alias) to the same underlying data.
* Any edits to df_copy also change df — which can cause major, hidden bugs for beginners.

>✔ df_copy = df.copy()
* This creates a true, independent copy.
* Editing one DataFrame will not affect the other.

**When to create a copy**
We create a copy when:

* ✔ Starting a new cleaning workflow

So we can compare before/after versions.

* ✔ Running experimental transformations

So mistakes don’t corrupt the original data.

* ✔ Preparing data for machine learning

Where training/validation/test splits must not modify the raw data.
